```python
# Install Tesseract OCR and Python packages
!apt-get update
!apt-get install -y tesseract-ocr tesseract-ocr-tam
!pip install pytesseract Pillow PyMuPDF

# Import required libraries
import pytesseract
from PIL import Image
import io
import fitz  # PyMuPDF
from google.colab import files

# Functions to extract text
def extract_text_from_image(image_file):
    image = Image.open(image_file)
    text = pytesseract.image_to_string(image, lang="tam+eng")
    return text.strip()

def extract_text_from_pdf(pdf_file):
    pdf_document = fitz.open(stream=pdf_file.read(), filetype="pdf")
    all_text = ""
    for page_num in range(len(pdf_document)):
        page = pdf_document[page_num]
        pix = page.get_pixmap()
        img_data = pix.tobytes("png")
        image = Image.open(io.BytesIO(img_data))
        text = pytesseract.image_to_string(image, lang="tam+eng")
        all_text += f"\n\n--- Page {page_num + 1} ---\n{text.strip()}"
    return all_text.strip()

# Upload files
uploaded = files.upload()  # Choose PDF or image files

# Process uploaded files
for filename in uploaded.keys():
    if filename.lower().endswith(".pdf"):
        with open(filename, "rb") as f:
            text = extract_text_from_pdf(f)
    else:
        text = extract_text_from_image(filename)

    print(f"Text from {filename}:\n{text}")

# Save extracted text as a file
with open("output_text.txt", "w", encoding="utf-8") as f:
    f.write(text)

files.download("output_text.txt")
```
